```python
import json

# 读取所有 JSON 数据到一个列表中
all_data = []
with open('Mental health Depression disorder Data.json', 'r', encoding='utf-8') as json_file:
    all_data = json.load(json_file)

# 提取 'Schizophrenia (%)' 和 'Entity' 的值到一个新的列表中
schizophrenia_data = []
for data_point in all_data:
    schizophrenia_percent_str = data_point.get('Schizophrenia (%)')
    country_name = data_point.get('Entity')
    
    # 检查值是否为有效的浮点数
    try:
        schizophrenia_percent = float(schizophrenia_percent_str)
    except (ValueError, TypeError):
        continue  # 跳过无效的值
    
    if country_name is not None:
        schizophrenia_data.append({'Country': country_name, 'Schizophrenia (%)': schizophrenia_percent})

# 根据 'Schizophrenia (%)' 从大到小排序
sorted_data = sorted(schizophrenia_data, key=lambda x: x['Schizophrenia (%)'], reverse=True)

# 打印排序后的国家数据
for item in sorted_data:
    print(f"国家: {item['Country']}, Schizophrenia (%): {item['Schizophrenia (%)']}")
```
