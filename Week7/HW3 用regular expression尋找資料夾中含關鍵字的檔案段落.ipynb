```python
import os
import pdfplumber
import re

# 定義要搜索的關鍵字，不限大小寫，可包含任何標點符號
keywords = [r'board game', r'game based learning', r'augmented reality']

# 定義PDF資料夾路徑
pdf_folder = r'C:\Users\ching\OneDrive - cycu.edu.tw\台師大\研究所論文\文獻'

# 查看所有資料夾中的PDF文件
for root, dirs, files in os.walk(pdf_folder):
    for file in files:
        if file.endswith('.pdf'):
            pdf_path = os.path.join(root, file)

            with pdfplumber.open(pdf_path) as pdf:
                for page in pdf.pages:
                    text = page.extract_text()
                    # 使用規則運算式將文本分割成段落
                    paragraphs = re.split(r'\n\s*\n', text)
                    for paragraph in paragraphs:
                        for keyword in keywords:
                            # 使用規則運算式搜索關鍵字並檢查段落是否包含關鍵字
                            if re.search(keyword, paragraph, re.IGNORECASE):
                                print(f"File: {file}, Page: {page.page_number}")
                                print(paragraph)
                                print("\n")
```

```python
import scholarly
import matplotlib.pyplot as plt

# 定义要搜索的关键字
keywords = ['board game', 'game based learning', 'augmented reality']

# 创建一个字典来存储年份-文献数量
publication_years = {}

# 遍历关键字并进行搜索
for keyword in keywords:
    search_query = scholarly.search_pubs_query(keyword)
    for i in range(10):  # 仅获取前10个文献
        try:
            publication = next(search_query)
            year = publication.bib.get('year', 'Unknown')
            if year != 'Unknown':
                year = int(year)
                if year in publication_years:
                    publication_years[year] += 1
                else:
                    publication_years[year] = 1
        except StopIteration:
            break

# 将数据整理成两个列表以便绘制趋势图
years = list(publication_years.keys())
counts = [publication_years[year] for year in years]

# 创建趋势图
plt.figure(figsize=(10, 6))
plt.plot(years, counts, marker='o')
plt.title('Publication Trends in Google Scholar')
plt.xlabel('Year')
plt.ylabel('Number of Publications')
plt.grid(True)
plt.legend(keywords)
plt.show()
```
