<font size="6">Q1.**日本相對於GDP最大和最小的國家，生育率如何?**<font size="6">

```python
import pandas as pd
import plotly.express as px

# 读取 GDP 数据
df1 = pd.read_csv('GDP.csv')

# 使用 melt 函数将 1960 到 2022 年度数据整合到 'Year' 列下
df1 = df1.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], 
               var_name='Year', 
               value_name='GDP')

# 将 NaN 值替换为 0
df1['GDP'].fillna(0, inplace=True)

# 读取 Fertility rate 数据
df2 = pd.read_csv('Fertility rate.csv')

# 使用 melt 函数将 1960 到 2022 年度数据整合到 'Year' 列下
df2 = df2.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], 
               var_name='Year', 
               value_name='Fertility Rate')

# 将 NaN 值替换为 0
df2['Fertility Rate'].fillna(0, inplace=True)

# 读取 Mental health 数据
df3 = pd.read_csv('Mental health Depression disorder Data.csv')

# 将所有 NaN 值替换为 0
df3.fillna(0, inplace=True)
```

```python
print(df1)
print(df2)
print(df3)
```

```python
# 计算每个国家的平均 GDP，并移除平均 GDP 为 0 的国家
average_gdp = df1[df1['GDP'] != 0].groupby('Country Name')['GDP'].mean().reset_index()

# 找到平均 GDP 最大和最小的国家
max_gdp_country = average_gdp.loc[average_gdp['GDP'].idxmax()]
min_gdp_country = average_gdp.loc[average_gdp['GDP'].idxmin()]
print("平均GDP最大的国家：" + max_gdp_country['Country Name'])
print("平均GDP最小的国家：" + min_gdp_country['Country Name'])
```

```python
#### 提取 Japan 和其他两个国家的生育率数据
japan_data = df2[df2['Country Name'] == 'Japan']
max_gdp_data = df2[df2['Country Name'] == max_gdp_country['Country Name']]
min_gdp_data = df2[df2['Country Name'] == min_gdp_country['Country Name']]


# 将 "Year" 列的数据类型转换为整数
merged_data['Year'] = merged_data['Year'].astype(int)

# 仅保留年份小于或等于2021的数据
merged_data = merged_data[merged_data['Year'] <= 2021]

# 使用 Plotly 绘制图表
fig = px.line(merged_data, x='Year', y='Fertility Rate', color='Country Name',
              title='Comparison of Fertility Rate for Japan, Max GDP, and Min GDP Countries')

# 显示图表
fig.show()

```

*結果顯示，相對於GDP最大和最小的國家，其生育率一直維持在3以下*

<font size="6">Q2.**日本相對於GDP最大和最小的國家失業率如何?**<font size="6">

```python
df4 = pd.read_csv('unemployment analysis.csv')
# 使用 melt 函数将 1960 到 2022 年度数据整合到 'Year' 列下
print(df4.head)
```

```python
df4 = df4.melt(id_vars=['Country Name', 'Country Code'], 
               var_name='Year', 
               value_name='Unemployment Rate')

# 将 NaN 值替换为 0
df4['Unemployment Rate'].fillna(0, inplace=True)
```

```python
# 選擇 Japan 和 Burundi 兩個國家的失業率數據
selected_countries = ['Japan', 'Burundi','Qatar']
filtered_data = df4[df4['Country Name'].isin(selected_countries)]

# 使用 Plotly Express 繪製小提琴圖
fig = px.violin(filtered_data, x='Country Name', y='Unemployment Rate', 
                box=True,  # 顯示內部箱形圖
                points='all',  # 顯示所有數據點
                title='Unemployment Rate Distribution for Japan and Burundi')
 
# 顯示圖表
fig.show()
```

*GDP最高的國家Burundi其失業率眾數在1-2之間，Japan在4.6-2.8之間，奇怪的是GDP最低的國家Qatar失業率竟然是最低的眾數在0.8-0.26之間*
