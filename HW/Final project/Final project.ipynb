```python
import googlemaps

# 替換為您的Google Maps API金鑰
api_key = 'AIzaSyDezjNpnZz1YkeNNXwtD9brJ9TYnzmbWP8'

# 創建Google Maps用戶端
gmaps = googlemaps.Client(key=api_key
                        )

# 輸入地點和關鍵字
location_input = input("請輸入地點（例如，大安區）：")
keyword_input = input("請輸入關鍵字（例如，餐廳）：")

# 使用Geocoding API查找經緯度
geocoding_result = gmaps.geocode(location_input, region='tw')  # 請根據需要更改地理區域
print(geocoding_result[0])
```

```python
all_places = []  # 儲存所有店家的列表

if geocoding_result:
    # 獲取第一個匹配的地點的經緯度
    location = (geocoding_result[0]['geometry']['location']['lat'], geocoding_result[0]['geometry']['location']['lng'])

    # 設置搜索半徑
    radius = 1200  # 搜索半徑（以米為單位）
        
    # 使用Places API進行搜索
    places = gmaps.places_nearby(location=location, keyword=keyword_input, radius=radius, language='zh-TW')
    
    # 計算店家數
    number = 0
    
    # 列印店家資訊並儲存到 all_places 列表中
    for place in places['results']:
        name = place['name']
        address = place.get('vicinity', '地址未提供')
        place_id = place['place_id']
        place_details = gmaps.place(place_id, language='zh-TW')
        opening_hours = place_details.get('result', {}).get('opening_hours', {'weekday_text': '未提供'})
        rating = place_details.get('result', {}).get('rating', '未提供')
        
        # 計算店家數
        number = number + 1

        # 儲存店家資訊到 all_places 列表
        all_places.append({
            '店名': name,
            '地址': address,
            '星级评价': rating,
            '營業時間': opening_hours['weekday_text'] if 'weekday_text' in opening_hours else '未提供'
        })
        # 清理星级评价數據，將無效值（例如"未提供"）設置為一個合適的默認值，例如0
        for place in all_places:
            rating = place.get('星级评价', '未提供')
            if rating == '未提供':
                place['星级评价'] = 0  # 或者使用一個適當的默認值
        # 將找到的店家按照星级评价（rating）大小排序
        sorted_places = sorted(all_places, key=lambda x: x['星级评价'], reverse=True)
               
        # 列印店家資訊
        for place in sorted_places:
            print(f'店名: {place["店名"]}')
            print(f'地址: {place["地址"]}')
            print(f'星级评价: {place["星级评价"]}')
            if 'weekday_text' in opening_hours:
                print('營業時間:')
                for day, hours in enumerate(opening_hours['weekday_text']):
                    print(f'  {day + 1}: {hours}')  # 為了從星期日開始，請將索引加1
            else:
                print('營業時間: 未提供')
#             print()
else:
    print("未找到地點的經緯度資訊。請檢查輸入的地點名稱。")
    
print("總共", number, "家")

# 將所有店家資訊存儲到一個檔案中（例如 JSON 檔案）
import json

with open('all_places.json', 'w', encoding='utf-8') as f:
    json.dump(all_places, f, ensure_ascii=False, indent=4)
```

```python
from datetime import datetime
import tkinter as tk
from tkinter import ttk
from tkcalendar import Calendar
from tkinter import messagebox

def get_selected_datetime():
    # 为变量设置默认值
    selected_start_hour, selected_start_minute = 0, 0
    selected_end_hour, selected_end_minute = 0, 0

    selected_date = cal.get_date()
    selected_day_of_week = cal.selection_get().strftime('%A')

    selected_start_time = start_time_picker.get()
    selected_end_time = end_time_picker.get()

    # 转换时间字符串为整数
    selected_start_hour, selected_start_minute = map(int, selected_start_time.split(':'))
    selected_end_hour, selected_end_minute = map(int, selected_end_time.split(':'))
    print('選擇時間:',selected_start_hour, selected_start_minute,'到', selected_end_hour, selected_end_minute)

    # 检查结束时间是否早于开始时间
    if (selected_end_hour, selected_end_minute) <= (selected_start_hour, selected_start_minute):
        messagebox.showerror('错误', '结束时间不能早于或等于开始时间')
        return

    # 创建包含所选日期和时间的字典
    selected_datetime_info = {
        'selected_date': selected_date,
        'selected_day_of_week': selected_day_of_week,
        'selected_start_time': selected_start_time,
        'selected_end_time': selected_end_time
    }
    
    # 在獲取星期的時候使用映射
    day_of_week_mapping = {
        'Monday': '星期一',
        'Tuesday': '星期二',
        'Wednesday': '星期三',
        'Thursday': '星期四',
        'Friday': '星期五',
        'Saturday': '星期六',
        'Sunday': '星期日'
    }
    selected_day_of_week_chinese = day_of_week_mapping[selected_datetime_info['selected_day_of_week']]
    
    
    # 打印选择后的时间
    print('Selected Datetime Information:', selected_day_of_week_chinese)

    # 关闭窗口
    root.destroy()

# 創建主窗口
root = tk.Tk()
root.title('日期和時間選擇器')

# 添加日期選擇器
cal = Calendar(root, selectmode="day", year=2023, month=11, day=22)
cal.pack()

# 添加開始時間選擇器
start_time_label = ttk.Label(root, text='選擇開始時間：')
start_time_label.pack()
time_options = [f'{hour:02d}:{minute:02d}' for hour in range(24) for minute in range(0, 60, 15)]
start_time_picker = ttk.Combobox(root, values=time_options)
start_time_picker.pack()

# 添加結束時間選擇器
end_time_label = ttk.Label(root, text='選擇結束時間：')
end_time_label.pack()
end_time_picker = ttk.Combobox(root, values=time_options)
end_time_picker.pack()

# 添加按鈕以獲取選擇的日期和時間
get_button = ttk.Button(root, text='獲取選擇的日期和時間', command=get_selected_datetime)
get_button.pack()

# 啟動主迴圈
root.mainloop()
```

```python
# 清理時間字符串的函數
def clean_time_string(time_str):
    return time_str.strip().lower()

# 筛选符合条件的店家
filtered_places = []
for place in all_places:
#     print(place)
    for opening_hours_entry in place.get('營業時間', []):
#         print(opening_hours_entry)

        # 转换时间字符串为整数
        if ':' in opening_hours_entry:
            day, hours_range_full = opening_hours_entry.split(':', 1)
            hours_range_full=clean_time_string(hours_range_full)
            # 检查是否休息
            if '休息' in hours_range_full or '未提供' in hours_range_full or ',' in hours_range_full:
#               print(f"{place['店名']} 在 {selected_day_of_week_chinese} 休息")
                continue  # 跳过该店家
            hours_range = hours_range_full.split('–')
#             print(hours_range)
            start_hour, start_minute = map(int, hours_range[0].split(':'))
            end_hour, end_minute = map(int, hours_range[1].split(':'))
#             print(day,start_hour, start_minute,end_hour, end_minute)

            # 检查是否符合条件
            if day == selected_day_of_week_chinese and  selected_start_hour <= start_hour and selected_end_hour <= end_hour:
                filtered_places.append(place)
        else:
            continue

# 打印符合条件的店家信息
if filtered_places:
    for place in filtered_places:
        print(f"店名: {place['店名']}")
        print(f"地址: {place['地址']}")
        print(f"星级评价: {place['星级评价']}")
        print(f"營業時間: {place['營業時間']}")
        print()
else:
    print("未找到符合条件的店家。")
```
